```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données
df = pd.read_csv('donnees.txt', sep='	')

# Renommer les colonnes pour un accès plus facile
df.columns = ['vitesse_vent', 'direction_vent', 'puissance_electrique', 'pression_atm', 'temperature', 'statut']

# Calculer la moyenne de la puissance électrique pour chaque vitesse de vent
moyenne_puissance = df.groupby('vitesse_vent')['puissance_electrique'].mean().reset_index()

# Créer le graphique
plt.figure(figsize=(12, 8))

# Afficher les données brutes
plt.scatter(df['vitesse_vent'], df['puissance_electrique'], label='Données brutes', alpha=0.5, s=10)

# Afficher la moyenne
plt.plot(moyenne_puissance['vitesse_vent'], moyenne_puissance['puissance_electrique'], color='red', linewidth=2, label='Moyenne de la puissance')

# Ajouter les titres et les légendes
plt.title('Puissance électrique en fonction de la vitesse du vent')
plt.xlabel('Vitesse du vent (m/s)')
plt.ylabel('Puissance électrique (W)')
plt.legend()
plt.grid(True)
plt.show()
```
